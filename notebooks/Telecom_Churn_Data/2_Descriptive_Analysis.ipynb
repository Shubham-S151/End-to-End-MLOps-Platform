# Descriptive Analysis on Churn Data

**Purpose:** Summarize and visualize the main characteristics of the data to understand distributions and key patterns.

This file contains descriptive analysis to uncover facts about the data and its components, list of task to do:
1. Summary statistics for numerical features (mean, std, min, max, 25th, 50th, 75th percentiles).
2. Histograms, boxplots, and density plots to inspect the distribution of individual features.
3. Count plots for categorical variables (frequencies of each category).
4. Correlation matrix (using heatmap) to understand relationships between numerical variables.
5. Pairwise scatter plots to inspect pairwise relationships between features.

```python
# importing necessary libraries
import os
import re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings("ignore")
```

```python
# creating data copy
data = df.copy()
data.head()
```

Insights: 
- There are columns which have all values as 0 like, `loc_og_t2o_mou`, `std_og_t2o_mou`, `loc_ic_t2o_mou` need to remove these redundant columns
- There are many columns with outliers
- Data have **null values** and most of them are present in **object datatype** columns
- Integer value columns might also be continous rather than sparse.
- Object datatype columns are date columns.
- Columns: `last_date_of_month_6`, `last_date_of_month_7`, `last_date_of_month_8`, `last_date_of_month_9` are not informative as they only give us the last date of month.

```python
# lets drop the last_date_of_month_*n columns
drop_cols = ["last_date_of_month_6", "last_date_of_month_7", "last_date_of_month_8", "last_date_of_month_9"]
data.drop(columns = drop_cols, inplace = True)
data.shape
```

```python
# creating a function to get the column name where there is only one value present
def get_col_with_one_val(data:pd.DataFrame)->tuple:
    return tuple(i for i in data.columns if data[i].nunique()<=1)

cols_with_one_vals = get_col_with_one_val(data)
```

```python
for i in cols_with_one_vals:
    print(f"Value in {i}: {data[i].unique()}")
```

We can drop these columns and it eill also help in removing the null values too.

```python
drop_cols.extend(cols_with_one_vals)
data.drop(columns = list(cols_with_one_vals), inplace=True)
data.shape
```

### Univariate Analysis

```python
# Making a seprate directory for storing plots
path = os.getcwd()
plots_dir = os.path.join(path, "plots")
univar_dir = os.path.join(plots_dir, "univariate_plots")

os.makedirs(plots_dir, exist_ok = True)
print(f"Directory created: {plots_dir}")

os.makedirs(univar_dir, exist_ok = True)
print(f"Directory created: {univar_dir}")
```

```python
non_null_cols = [col for col in data.columns if data[col].isna().sum()!=0]
print("Count of non null columns: ",len(non_null_cols))
```

```python
nn_nuniq_10less = {col:data[col].nunique() for col in non_null_cols if data[col].nunique()<=10}
nn_nuniq_10less
```

```python
cat_cols = [col for col in non_null_cols if data[col].dtype == object]
print(cat_cols)
```

```python
print("Count of non null continous columns: ",len(non_null_cols)-len(nn_nuniq_10less.keys())-len(cat_cols))
```

```python
last_rech_date_cols = ["date_of_last_rech_"+str(i) for i in range(6,10)]
print(last_rech_date_cols)
```

```python
# we can make group of 12 columns and make a combined plot for them
cont_cols = [i for i in non_null_cols if i not in cat_cols and i not in last_rech_date_cols and i not in nn_nuniq_10less.keys()]
print(len(cont_cols))
chunk_size = 9   # 9 plots per figure
group_cols = [cont_cols[i:i+chunk_size] for i in range(0, len(cont_cols), chunk_size)]
print(group_cols[0])
```

```python
# making grouped kde-plots
# creating the subfolder for kde plot
kde_plots_dir = os.path.join(univar_dir,"KDE_Plots")
os.makedirs(kde_plots_dir, exist_ok = True)
print(f"Directory created: {kde_plots_dir}")

rows, cols = 3, 3
for i, group in enumerate(group_cols):
    fig, axes = plt.subplots(rows, cols, figsize=(18, 18))
    axes = axes.flatten()
    for j, col in enumerate(group):
        sns.kdeplot(data=data, x=col, fill=True, ax=axes[j])
        mean_val = data[col].mean()
        median_val = data[col].median()
        mode_val = data[col].mode().iloc[0]
        # Add vertical lines
        axes[j].axvline(mean_val, color="red", linestyle="--", label=f"Mean = {mean_val:.2f}")
        axes[j].axvline(median_val, color="blue", linestyle="--", label=f"Median = {median_val:.2f}")
        axes[j].axvline(mode_val, color="green", linestyle="--", label=f"Mode = {mode_val:.2f}")
        # Add legend (inside each subplot)
        axes[j].legend(fontsize=9)
        # Title with column name
        axes[j].set_title(col)
    # Hide unused axes if last group < 9
    for k in range(len(group), len(axes)):
        axes[k].set_visible(False)
    plt.suptitle(f"KDE Plots - Group {i+1}", fontsize=22)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    # Save figure
    fig_path = os.path.join(kde_plots_dir, f"KDE_Group_{i+1}.png")
    plt.savefig(fig_path)
    print(f"KDE_Group_{i+1} saved...")
    plt.close(fig)   # Close to free memory

print(f"All KDE plots saved in: {kde_plots_dir}")
```

```python
# making dir and plots for categorical columns
# cerating barplot dir
bar_plots_dir = os.path.join(univar_dir,"Bar_Plots")
os.makedirs(bar_plots_dir, exist_ok = True)
print(f"Directory created: {bar_plots_dir}")

rows, cols = 3, 3
group_cols = [cat_cols[i:i+chunk_size] for i in range(0, len(cat_cols), chunk_size)]

# --- Plotting Barplots ---
for i, group in enumerate(group_cols):
    fig, axes = plt.subplots(rows, cols, figsize=(18, 18))
    axes = axes.flatten()
    
    for j, col in enumerate(group):
        # Count the unique values in the categorical column
        value_counts = data[col].value_counts()
        
        # Plot the bar plot for categorical column
        sns.barplot(x=value_counts.index, y=value_counts.values, ax=axes[j], palette="Set2")
        
        # Title and labels
        axes[j].set_title(f"{col}")
        axes[j].set_xlabel(col)
        axes[j].set_ylabel("Count")
        
        # Rotate x-ticks for better readability
        for label in axes[j].get_xticklabels():
            label.set_rotation(45)
        
    # Hide unused axes if the last group has less than 9 categories
    for k in range(len(group), len(axes)):
        axes[k].set_visible(False)
    
    plt.suptitle(f"Categorical Bar Plots - Group {i+1}", fontsize=18)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    
    # Save the figure to the directory
    fig_path = os.path.join(bar_plots_dir, f"Cat_Group_{i+1}.png")
    plt.savefig(fig_path)
    plt.close(fig)   # Close to free memory

print(f"All categorical bar plots saved in: {bar_plots_dir}")
```
