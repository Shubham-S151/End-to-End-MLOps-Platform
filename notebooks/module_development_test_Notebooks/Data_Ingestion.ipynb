```python
# import libraries
import os
import sys
import pandas as pd
```

```python
def map_type(file_type: str) -> str:
        fmap = {
            # Tabular data
            "csv": "csv",
            "tsv": "csv",
            "txt": "text",

            # Excel
            "xls": "excel",
            "xlsx": "excel",
            "xlsm": "excel",

            # JSON / semi-structured
            "json": "json",

            # Columnar data (big data / ML)
            "parquet": "parquet",
            "feather": "feather",
        }
        return fmap.get(file_type, "unknown")
```

```python
def check_file_type(data:str)->str:
    """This is a function to check and retrive the file type which then can be used for loading data.
    Params:
        data: path of the data file. (for ex: C:/-----/<file_name>)
    Results:
        Returns the file type str like: excel, csv, etc.
    """
    if not data:
        raise ValueError("File path must be provided")
    
    if os.path.isdir(data):
        raise ValueError("Folder path is passed instead of a file")
    
    file_name = os.path.basename(data.lower())
    if "." not in file_name:
        raise ValueError("File has no extension")
    
    ftype = file_name.rsplit(".", 1)[-1]
    return map_type(ftype)

check_file_type(r"D:\Doccuments\GitHub\End-to-End-MLOps-Platform\data\raw\fraud test.csv")
```
